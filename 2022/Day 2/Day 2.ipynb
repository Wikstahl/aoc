```python
def calc_total_score(input_file):
  # Define a dictionary that maps each hand shape to the hand shape that beats it
  beaten_by = {"A": "Y", "B": "X", "C": "Z"}

  # Define a dictionary that maps each hand shape to its score
  score_of = {"A": 1, "B": 2, "C": 3}

  # Read the strategy guide from the input file
  with open(input_file) as f:
    strategy = f.read().strip()
  
  # Split the strategy into rounds
  rounds = strategy.split("\n")

  # Initialize the total score to 0
  total_score = 0

  # Iterate through each round
  for round in rounds:
    # Split the round into the opponent's hand shape and our hand shape
    opponent, me = round.split()
    print(opponent,me)
    # If we win, our score is the score of the hand shape that we chose plus 6
    if beaten_by[opponent] == me:
      total_score += score_of[me] + 6

    # If we lose, our score is the score of the hand shape that we chose plus 0
    elif opponent == beaten_by[me]:
      total_score += score_of[me]

    # If we draw, our score is the score of the hand shape that we chose plus 3
    else:
      total_score += score_of[me] + 3

  # Return the total score
  return total_score
calc_total_score('input.txt')
```
