```python
data = open("data.txt").read().splitlines()
```

```python
data = list(map(int, data))
```

```python
data
```

## Part 1

```python
for i in range(len(data)):
    for j in range(i+1,len(data)):
        value = data[i] + data[j]
        if value == 2020:
            print(data[i]*data[j])
            break
```

## Part 2

```python
for i in range(len(data)):
    for j in range(i+1,len(data)):
        for k in range(j+1, len(data)):
            value = data[i] + data[j] + data[k]
            if value == 2020:
                print(data[i]*data[j]*data[k])
                break
```
